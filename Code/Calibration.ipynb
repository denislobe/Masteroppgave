```python
%matplotlib qt5
import hyperspy.api as hs
import numpy as np
import pyxem as pxm
from pyxem.libraries.calibration_library import CalibrationDataLibrary
from pyxem.generators.calibration_generator import CalibrationGenerator
from pathlib import Path
```

```python
SPED5_path = Path("D:/datasets/250225_Ag_oxww/20250225_175525/SPED5.zspy")
SPED5 = hs.load(SPED5_path, lazy=True)
nav_y, nav_x = SPED5.axes_manager.navigation_axes
nav_y.units, nav_x.units = "nm", "nm"
nav_y.scale = 1.745
nav_x.scale = 1.745
SPED5.calibrate()
```

```python
s_shifts = SPED5.get_direct_beam_position(method="blur", sigma=5, half_square_width=20)
s_shifts.compute()
linear_shifts = s_shifts.get_linear_plane()
SPED5_calibrated = SPED5.center_direct_beam(shifts=linear_shifts, inplace=False)
SPED5_calibrated.save("D:/datasets/250225_Ag_oxww/20250225_175525/SPED5_calibrated.zspy")

sig1_y, sig1_x = SPED5.axes_manager.signal_axes
```
